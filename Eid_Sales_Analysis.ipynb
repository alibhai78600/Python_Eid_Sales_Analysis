# Exploratory Data Analysis

### Gender

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(rc={'figure.figsize':(6,5)})

ax = sns.countplot(x = 'Gender', data = df, hue = 'Gender', palette = 'Set2', legend = False)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data grouping aur sorting
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})

ax = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen, hue = 'Gender', palette = 'magma', legend = False)

for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

*From above graphs we can see that most of the buyers are males and even the purchasing power of males are greater than female*

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data grouping aur sorting
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(10,5)})

ax = sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age, hue = 'Age Group', palette = 'viridis', legend = False)

for i in ax.containers:
    ax.bar_label(i, fmt='%.0f', padding=3)

plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

### State

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data grouping aur sorting (Top 10 States by Orders)
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

ax = sns.barplot(data = sales_state, x = 'State', y = 'Orders', hue = 'State', palette = 'magma', legend = False)

for i in ax.containers:
    ax.bar_label(i)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data grouping aur sorting (Top 10 States)
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

ax = sns.barplot(data = sales_state, x = 'State', y = 'Amount', hue = 'State', palette = 'viridis', legend = False)

for i in ax.containers:
    ax.bar_label(i, fmt='%.0f', padding=3)

plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

*From above graphs we can see that most of the orders & total sales/amount are from Makkah, New York and Mumbai respectively*


### Marital Status

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(rc={'figure.figsize':(7,5)})

ax = sns.countplot(data = df, x = 'Marital_Status', hue = 'Marital_Status', palette = 'Set2', legend = False)

for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

### Occupation

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(data = df, x = 'Occupation', hue = 'Occupation', palette = 'crest', legend = False)

for bars in ax.containers:
    ax.bar_label(bars)

plt.xticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data grouping aur sorting
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)


sns.set(rc={'figure.figsize':(20,5)})

ax = sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount', hue = 'Occupation', palette = 'flare', legend = False)

for i in ax.containers:
    ax.bar_label(i, fmt='%.0f', padding=3)

plt.ticklabel_format(style='plain', axis='y')
    
plt.xticks(rotation=45)

plt.show()
```

*From above graphs we can see that most of the buyers are working in Manager and sales sector*

### Product Category

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Figure size aur style set karna
sns.set(rc={'figure.figsize':(20,5)})

# 1. 'palette' se color change hoga
# 2. 'hue' use karne se warning nahi aayegi
# 3. 'palette' ki jagah aap 'viridis', 'flare', 'crest', ya 'rocket' likh sakte hain
ax = sns.countplot(data = df, x = 'Product_Category', hue = 'Product_Category', palette = 'viridis', legend = False)

# Bars ke upar numbers (counts) dikhane ke liye
for bars in ax.containers:
    ax.bar_label(bars)

# Categories ke naam agar bade hain toh unhe thoda tedha (rotate) karne ke liye taaki wo clear dikhen
plt.xticks(rotation=45)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Data grouping aur sorting (Top 10 categories by Amount)
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})

# 3. Barplot with Colors
ax = sns.barplot(data = sales_state, x = 'Product_Category', y = 'Amount', hue = 'Product_Category', palette = 'magma', legend = False)

for i in ax.containers:
    ax.bar_label(i, fmt='%.0f', padding=3)

plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

*From above graphs we can see that most of the sold products are from Groceries,Beauty,and Clothing category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders', hue = 'Product_ID', palette = 'viridis', legend = False)
```

```python
import seaborn as sns

plt.figure(figsize=(12,7))

# Top 10 data nikalein
sales_data = df.groupby('Product_ID')['Orders'].sum().nlargest(10).reset_index()

# Seaborn barplot (Warning hatane ke liye hue aur legend add kiya hai)
ax = sns.barplot(x='Product_ID', y='Orders', data=sales_data, palette='viridis', hue='Product_ID', legend=False)

# Bars ke upar value dikhane ke liye
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```
